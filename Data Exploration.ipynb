# USA Per Capita Overdose Rate by Year

Age-adjusted death rates (deaths per 100,000 U.S. standard population for 2000) are calculated using the direct method

```python
NCHS.loc[(NCHS['State'] == 'United States') 
         & (NCHS['Age Group'] == 'All Ages') 
         & (NCHS['Sex'] == 'Both Sexes')
        & (NCHS['Race and Hispanic Origin'] == 'All Races-All Origins')][
    ['Year','US Age-adjusted Rate']].set_index('Year')
```
