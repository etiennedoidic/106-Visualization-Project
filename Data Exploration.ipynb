# USA Per Capita Overdose Rate by Year

Age-adjusted death rates (deaths per 100,000 U.S. standard population for 2000) are calculated using the direct method

```python
line_chart_data = NCHS.loc[(NCHS['State'] == 'United States') 
         & (NCHS['Age Group'] == 'All Ages') 
         & (NCHS['Sex'] == 'Both Sexes')
        & (NCHS['Race and Hispanic Origin'] == 'All Races-All Origins')][
    ['Year','US Age-adjusted Rate']].set_index('Year')
line_chart_data
```

```python
data = {}
data['year'] = list(line_chart_data.index)
data['age_adjusted_rate'] = list(line_chart_data['US Age-adjusted Rate'])
with open('Data/line_chart_data.json', 'w') as outfile:
    json.dump(data, outfile)
```
