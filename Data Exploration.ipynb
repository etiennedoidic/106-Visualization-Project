# USA Per Capita Overdose Rate by Year

Age-adjusted death rates (deaths per 100,000 U.S. standard population for 2000) are calculated using the direct method

```python
line_chart_data = NCHS.loc[(NCHS['State'] == 'United States') 
         & (NCHS['Age Group'] == 'All Ages') 
         & (NCHS['Sex'] == 'Both Sexes')
        & (NCHS['Race and Hispanic Origin'] == 'All Races-All Origins')][
    ['Year','US Age-adjusted Rate']].set_index('Year')
line_chart_data
```

```python
data = {}
data['year'] = list(line_chart_data.index)
data['age_adjusted_rate'] = list(line_chart_data['US Age-adjusted Rate'])
with open('Data/line_chart_data.json', 'w') as outfile:
    json.dump(data, outfile)
```

```python
NCHS.loc[(NCHS['State'] == 'United States') 
         & (NCHS['Age Group'] == 'All Ages') 
         & (NCHS['Sex'] == 'Both Sexes')
        & (NCHS['Race and Hispanic Origin'] == 'All Races-All Origins')][
    ['Year','Deaths', 'Population']].set_index('Year')
```

```python
#collects the state year and rate of the data I will be converting to a json file
state_data = NCHS.drop(NCHS.loc[(NCHS['State'] == 'United States') | (NCHS['State'] == 
                        'District of Columbia')].index)[['State', 'Year', 'Age-adjusted Rate']]
```

```python
#makes the dictionary that I will push the state's rate to
years = state_data['Year'].unique()
states = state_data['State'].unique()
states_dict = {}
for s in states:
    states_dict[s] = []
    for y in years:
        states_dict[s].append(round(state_data.loc[(state_data['Year'] == y) 
                        & (state_data['State'] == s)]['Age-adjusted Rate'].values[0], 4))
```

```python
years
```

```python
#pushes the dictionary to a json file
with open('Data/states_data.json', 'w') as outfile:
    json.dump(states_dict, outfile)
```

```python
VSRR = pd.read_csv('Data/VSRR_Provisional_Drug_Overdose_Death_Counts.csv')
```

```python

list(VSRR.Indicator.unique())
```

```python
toExport = {}
for state in VSRR.State.unique():
    currdf = VSRR.loc[(VSRR['State'] == state) & (VSRR['Year'] == 2018)].groupby('Indicator').sum()
    state_info = {}
    for drug in currdf.index:
        state_info[drug] = currdf.loc[currdf.index == drug]['Data Value'].values[0]
    toExport[state] = state_info
toExport
```

```python
with open('Data/pieChartData.json', 'w') as outfile:
    json.dump(toExport, outfile)
```
